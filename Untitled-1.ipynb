Identificando os formatos: JPG, tif, jpg e png

```python
from PIL import Image
import numpy as np
desired_shape = (248, 248)
imagens =[]
for arquivo in os.listdir(pasta):
    if arquivo.endswith(".png") or arquivo.endswith(".JPG") or arquivo.endswith(".tif") or arquivo.endswith(".jpg"):
        caminho = os.path.join(pasta, arquivo)
        imagem = Image.open(caminho)
        imagem = imagem.resize(desired_shape)
        imagens.append(np.array(imagem))
```

Pré-Processamento

Técnicas possíveis:

*Normalização: A normalização é aplicada para garantir que os dados de entrada tenham uma distribuição uniforme, o que pode ajudar a evitar problemas como a saturação de pixels brilhantes ou pixels escuros.

*Filtragem: A filtragem é aplicada para remover ruído ou outras características indesejadas das imagens de entrada.

*Reescalonamento: O reescalonamento é aplicado para ajustar o tamanho das imagens de entrada para que sejam todas do mesmo tamanho.

Obs: as técnicas devem ser empregadas nessa ordem

```python
images = []
for imagem in imagens:
    # Carregar a imagem
    img = cv2.imread(imagem)
    # Verificar se a imagem foi carregada corretamente
    if img is None:
        print("Não foi possível carregar a imagem", imagem)
    else:
        images.append(img)
```

```python
#Colocar as imagens em escala de cinza
if imagens is None:
    raise ValueError("Não foi possível carregar a imagem")
gray = cv2.cvtColor(imagens, cv2.COLOR_BGR2GRAY)
#Plotar histograma
hist = cv2.calcHist([gray],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```
